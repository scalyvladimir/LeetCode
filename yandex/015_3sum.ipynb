## 15. 3Sum

```python
def partition(l, r, arr):
    pivot = arr[r-1]
    i = l - 1
    for j in range(l, r):
        if arr[j] <= pivot:
            i += 1
            if i != j:
                arr[i], arr[j] = arr[j], arr[i]
    return i

def qSort(l, r, arr):
    if l < r:
        p = partition(l, r, arr)
        qSort(l, p, arr)
        qSort(p+1, r, arr)
```

```python
def threeSum(nums):
    qSort(0, len(nums), nums)
    
    ans = []

    for el in range(len(nums)-2):
        if el > 0 and nums[el] == nums[el-1]:
            continue

        beg = el + 1
        end = len(nums)-1
        while beg < end:
            sum = nums[beg] + nums[el] + nums[end]
            if sum < 0:
                beg += 1
            elif sum > 0:
                end -= 1
            else:
                ans.append([nums[beg], nums[el], nums[end]])
                while beg < end and nums[beg] == nums[beg+1]:
                    beg += 1
                while beg < end and nums[end] == nums[end-1]:
                    end -= 1
                
                beg += 1
                end -= 1

    return ans
```
